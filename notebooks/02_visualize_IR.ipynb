# 02 — Visualize and Explore IR
This notebook loads the Intermediate Representation (IR) generated in Notebook 01
and provides tools to inspect, validate, and visualize its structure.

The goal is to understand how IR acts as the stable, deterministic layer between
natural-language test cases and automated test generation.


```python
import json
from pathlib import Path
from pprint import pprint
```

## 1. Load the IR File

We load the IR JSON saved by Notebook 01.


```python
ir_path = Path("../sample-data/ir_examples/login_ir.json")

with open(ir_path, "r") as f:
    ir = json.load(f)

pprint(ir)
```

## 2. Inspect IR Structure

We examine the top-level keys and the structure of the `steps` list.


```python
print("Top-level keys:", list(ir.keys()))
print("\nSteps:")
for step in ir["steps"]:
    print(" -", step)
```

## 3. Visualize IR as a Tree

A simple tree-style visualization helps understand the hierarchy.


```python
def print_tree(obj, indent=0):
    prefix = " " * indent
    if isinstance(obj, dict):
        for k, v in obj.items():
            print(f"{prefix}- {k}:")
            print_tree(v, indent + 4)
    elif isinstance(obj, list):
        for i, item in enumerate(obj):
            print(f"{prefix}- [{i}]")
            print_tree(item, indent + 4)
    else:
        print(f"{prefix}{obj}")

print_tree(ir)
```

## 4. Validate IR Fields

A simplified validation step to ensure required fields exist.


```python
required_top_level = ["test_name", "steps"]

missing = [k for k in required_top_level if k not in ir]
if missing:
    print("Missing required fields:", missing)
else:
    print("IR contains all required top-level fields.")
```

## 5. Validate Step Structure

Each step should contain at least an `action` and a `target`.


```python
invalid_steps = []

for i, step in enumerate(ir["steps"]):
    if "action" not in step or "target" not in step:
        invalid_steps.append((i, step))

if invalid_steps:
    print("Invalid steps found:")
    for idx, step in invalid_steps:
        print(f" - Step {idx}: {step}")
else:
    print("All steps contain required fields.")
```

## 6. Export a Pretty-Printed Version

This is useful for documentation or debugging.


```python
pretty_path = Path("../sample-data/ir_examples/login_ir_pretty.json")

with open(pretty_path, "w") as f:
    json.dump(ir, f, indent=4)

pretty_path
```
