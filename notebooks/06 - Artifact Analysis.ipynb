# 06 — Artifact Analysis

This notebook loads the artifacts produced by Notebook 05 and generates a
human-readable analysis of the test run.

Artifacts include:
- summary.json
- execution_log.txt

Pipeline:
NL → IR → Mapped IR → Code → Execution → Analysis


```python
from pathlib import Path
import json
```

## 1. Locate the Most Recent Run Directory


```python
from pathlib import Path

# Point to the new Flask app artifacts root
runs_root = Path("../artifacts/flask_app_run")

run_dirs = sorted(runs_root.glob("run_*"))
latest_run = run_dirs[-1] if run_dirs else None

latest_run
```

## 2. Load Summary JSON


```python
summary_path = latest_run / "summary.json"

with open(summary_path, "r") as f:
    summary = json.load(f)

summary
```

## 3. Load Execution Log


```python
log_path = latest_run / "execution_log.txt"

with open(log_path, "r") as f:
    log_text = f.read()

log_text
```

## 4. Analyze the Run

We extract:
- Pass/Fail
- Errors (if any)
- Key signals from stdout/stderr


```python
analysis = {}

# Pass/fail
analysis["passed"] = (summary["return_code"] == 0)

# Errors
if summary["stderr"]:
    analysis["error"] = summary["stderr"]
else:
    analysis["error"] = None

# Stdout
analysis["stdout"] = summary["stdout"] if summary["stdout"] else "(no stdout)"

# Stderr
analysis["stderr"] = summary["stderr"] if summary["stderr"] else "(no stderr)"

analysis
```

## 5. Produce a Human-Readable Report


```python
def render_report(analysis, summary, log_text):
    lines = []
    lines.append("=== Test Execution Report ===")
    lines.append(f"Test File: {summary['test_file']}")
    lines.append(f"Artifacts Directory: {summary['artifacts_dir']}")
    lines.append("")
    lines.append(f"Result: {'PASS' if analysis['passed'] else 'FAIL'}")
    lines.append("")
    lines.append("=== Stdout ===")
    lines.append(analysis["stdout"])
    lines.append("")
    lines.append("=== Stderr ===")
    lines.append(analysis["stderr"])
    lines.append("")
    lines.append("=== Raw Execution Log ===")
    lines.append(log_text)
    return "\n".join(lines)

report = render_report(analysis, summary, log_text)
print(report)
```
