# 01 — Parse Test Case to IR
This notebook demonstrates how a natural-language test case can be transformed into a structured Intermediate Representation (IR).
The IR is the foundation of the deterministic test automation pipeline.


```python
import json
from pathlib import Path
```

## 1. Load a Sample Test Case

We start with a simple, human-written test case stored in `sample-data/test_cases/`.
This simulates a Confluence-style manual test case.


```python
sample_path = Path("../sample-data/test_cases/login_test_case.txt")

with open(sample_path, "r") as f:
    test_case_text = f.read()

print(test_case_text)
```

## 2. Convert Test Case to IR (Prototype)

This cell simulates the IR generation step.

In the private engine, this would be handled by:
- an LLM prompt
- a deterministic schema validator
- normalization logic

For the public demo, we mock the output.


```python
# Mock IR output for demo purposes
ir = {
    "test_name": "login_valid_credentials",
    "description": "User logs in with valid credentials",
    "steps": [
        {"action": "navigate", "target": "login_page"},
        {"action": "input", "target": "username_field", "value": "valid_user"},
        {"action": "input", "target": "password_field", "value": "valid_password"},
        {"action": "click", "target": "login_button"},
        {"action": "assert", "target": "dashboard_page", "condition": "visible"}
    ],
    "metadata": {
        "priority": "high",
        "tags": ["smoke", "auth"]
    }
}

ir
```

## 3. Visualize IR

This makes the IR easy to inspect and compare.


```python
print(json.dumps(ir, indent=4))
```

## 4. Validate IR Structure (Simplified)

In the private engine, this would use:
- a strict schema
- type checking
- field validation
- error reporting

Here we demonstrate a simple structural check.


```python
required_keys = ["test_name", "steps"]

missing = [k for k in required_keys if k not in ir]
missing
```

## 5. Save IR Example

This writes the IR to `sample-data/ir_examples/` so other notebooks can use it.


```python
output_path = Path("../sample-data/ir_examples/login_ir.json")
output_path.parent.mkdir(parents=True, exist_ok=True)

with open(output_path, "w") as f:
    json.dump(ir, f, indent=4)

output_path
```
