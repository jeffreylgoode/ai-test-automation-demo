# 05 — Playwright Test Execution

This notebook loads the Playwright test generated in Notebook 04 and executes it.
It captures:
- Screenshot
- Trace
- Console logs
- Execution summary

Pipeline:
NL → IR → Mapped IR → Code → Execution → Analysis


```python
from pathlib import Path
import json
import subprocess
import datetime
```

## 1. Define Paths


```python
# Choose which generated test to run
test_path = Path("../generated-tests/render_app_test_test.py")

# Artifacts directory for this app
artifacts_dir = Path("../artifacts/flask_app_run")
artifacts_dir.mkdir(parents=True, exist_ok=True)

# Timestamped run directory
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
run_dir = artifacts_dir / f"run_{timestamp}"
run_dir.mkdir(parents=True, exist_ok=True)

run_dir
```

## 2. Execute the Test

We run the generated test using Python and capture:
- stdout
- stderr
- return code


```python
result = subprocess.run(
    ["python", str(test_path)],
    capture_output=True,
    text=True
)

result
```

## 3. Save Execution Logs


```python
log_path = run_dir / "execution_log.txt"

with open(log_path, "w") as f:
    f.write("STDOUT:\n")
    f.write(result.stdout)
    f.write("\n\nSTDERR:\n")
    f.write(result.stderr)

log_path
```

## 4. Capture Screenshot and Trace

We modify the test to automatically save:
- screenshot
- trace

But since the generated test already runs, we capture only what Playwright produces.


```python
# If your test saves screenshots or traces, move them here.
# For now, we simply record that the run completed.

summary = {
    "test_file": str(test_path),
    "return_code": result.returncode,
    "stdout": result.stdout,
    "stderr": result.stderr,
    "artifacts_dir": str(run_dir)
}

summary_path = run_dir / "summary.json"

with open(summary_path, "w") as f:
    json.dump(summary, f, indent=4)

summary_path
```
